```python
#Value Counts
#Donâ€™t forget the method "value_counts" only works on pandas series, not pandas dataframes. As a result, we only include one bracket df['drive-wheels'], not two brackets df[['drive-wheels']]
df["drive-wheels"].value_counts()
```

```python
#We can convert the series to a dataframe as follows:
df["drive-wheels"].value_counts().to_frame()
```

```python
drive_wheels_counts = df["drive-wheels"].value_counts().to_frame()
drive_wheels_counts.reset_index(inplace=True)
drive_wheels_counts = drive_wheels_counts.rename(columns={"drive-wheels":"drive_wheels_counts"})
drive_wheels_counts
```

```python
# engine-location as variable
engine_locations_counts = df["engine-location"].value_counts().to_frame()
engine_locations_counts.reset_index(inplace=True)
engine_locations_counts = engine_locations_counts.rename(columns={"engine-location":"engine_locations_counts"})
engine_locations_counts
```

```python
#Group by()
df_group_one = df[["drive-wheels", "body-style", "price"]]
df_grouping = df_group_one.groupby(["drive-wheels", "body-style"], as_index=False).mean()
df_grouping
```

```python
#visualizing using a pivot table
df_pivot = df_grouping.pivot(index="drive-wheels", columns ="body-style")
df_pivot
```

```python
df_pivot = df_pivot.fillna(0)
df_pivot
```

```python
df_group_body_style = df[["body-style", "price"]]
df_grouping_body_style = df_group_body_style.groupby(["body-style"], as_index=False).mean()
df_grouping_body_style
```

```python
#Let's use a heat map to visualize the relationship between Body Style vs Price.
plt.pcolor(df_pivot, cmap="RdBu")
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(df_pivot, cmap="RdBu")

#label names
row_labels = df_pivot.columns.levels[1]
col_labels = df_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(df_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(df_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

```python
#P-value
from scipy import stats
```

```python
#Let's calculate the Pearson Correlation Coefficient and P-value of 'wheel-base' and 'price'.
pearson_coef, p_value = stats.pearsonr(df["wheel-base"], df["price"])
print("The pearson correlation coefficient is", pearson_coef, "with pearson value of", p_value)
```

```python
#Let's calculate the  Pearson Correlation Coefficient and P-value of 'horsepower' and 'price'.
pearson_coef, p_value = stats.pearsonr(df["horsepower"], df["price"])
print("The pearson correlation coefficient is", pearson_coef, "with pearson value of", p_value)
```
