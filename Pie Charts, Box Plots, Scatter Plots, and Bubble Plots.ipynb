```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df_can = pd.read_csv(r"F:\Data_Science_Jupyter_Files\Canada.csv")
```

```python
# Visualizing Data using Matplotlib
```

```python
df_can.set_index("Country", inplace=True)
df_can.head()
```

```python
# let's create a list of years from 1980 - 2013, this will come in handy when we start plotting the data
years = list(map(str, range(1980,2014)))
```

```python
df_can.drop(["Region", "DevName"], axis=1, inplace=True)
df_can.head()
```

```python
# Pie Charts

# group countries by continents and apply sum() function
df_continents = df_can.groupby("Continent", axis=0).sum()

# note: the output of the groupby method is a `groupby' object. 
# we can not use it further until we apply a function (eg .sum())
print(type(df_can.groupby("Continent", axis=0)))

df_continents.head()
```

```python
# autopct create %, start angle represent starting point
df_continents["Total"].plot(kind="pie",
                           figsize=(12,6),
                           autopct="%1.1f%%",  # add in percentages
                           startangle=90,  # start angle 90° (Africa)
                           shadow=True,  # add shadow
                           )
plt.title("Immigration to Canada by Continent")
plt.axis("equal")
plt.legend(labels=df_continents.index, loc="upper left")
plt.show()
```

```python
# The above visual is not very clear, the numbers and text overlap in some instances. Let's make a few modifications to improve the visuals:
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1]

df_continents["Total"].plot(kind="pie",
                            figsize=(12, 6),
                           autopct="%1.1f%%",
                           labels=None,
                           startangle=90,
                           shadow=True,
                           explode=explode_list,
                            colors=colors_list,
                           pctdistance=1.12)
# scale the title up by 12% to match pctdistance
plt.title("Immigration to Canada by Continent", y=1.12, fontsize=15)
plt.axis("equal")
# add legend
plt.legend(labels=df_continents.index, loc="upper left", fontsize=7)
plt.show()
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1]

df_continents["2013"].plot(kind="pie",
                            figsize=(12, 6),
                           autopct="%1.1f%%",
                           labels=None,
                           startangle=90,
                           shadow=True,
                           explode=explode_list,
                            colors=colors_list,
                           pctdistance=1.12)
# scale the title up by 12% to match pctdistance
plt.title("Immigration to Canada by Continent", y=1.12, fontsize=15)
plt.axis("equal")
# add legend
plt.legend(labels=df_continents.index, loc="upper left", fontsize=7)
plt.show()
```

```python
# Question: Create a box plot to visualize the distribution of the top 15 countries (based on total immigration) grouped by the decades 1980s, 1990s, and 2000s.

df_top15 = df_can.sort_values("Total", ascending=False, axis=0)
df_top15 = df_top15.head(15)
df_top15
```

```python
# Create a new dataframe which contains the aggregate for each decade.

# create a list of all years in decades 80's, 90's, and 00's
years_80s = list(map(str, range(1980, 1990)))
years_90s = list(map(str, range(1990, 2000)))
years_00s = list(map(str, range(2000, 2010)))

# slice the original dataframe df_can to create a series for each decade
df_80s = df_top15.loc[:, years_80s].sum(axis=1)
df_90s = df_top15.loc[:, years_90s].sum(axis=1)
df_00s = df_top15.loc[:, years_00s].sum(axis=1)

# merge the three series into a new data frame
new_df = pd.DataFrame({"1980s": df_80s, "1990s": df_90s, "2000s": df_00s})
new_df.head()
```

```python
new_df.describe()
```

```python
new_df.plot(kind="box", figsize=(12,6))
plt.title('Immigration from top 15 countries for decades 80s, 90s and 2000s')
plt.show()
```

```python
# Scatter Plots 

# we can use the sum() method to get the total population per year
df_tot = pd.DataFrame(df_can[years].sum(axis=0))

# change the years to type int (useful for regression later on)
df_tot.index = map(int, df_tot.index)

# reset the index to put in back in as a column in the df_tot dataframe
df_tot.reset_index(inplace=True)

# rename columns
df_tot.columns = ["year", "total"]

# view the final dataframe
df_tot.head()
```

```python
# Step 2: Plot the data

df_tot.plot(kind="scatter", x="year", y="total", figsize=(12,6), color='darkgreen')
plt.title("Total Immigration to Canada from 1980 - 2013")
plt.xlabel("Year")
plt.ylabel("Number of Immigrants")
plt.show()
```

```python
# We can mathematically analyze this upward trend using a regression line (line of best fit)

# plot a linear line of best fit, and use it to predict the number of immigrants in 2015.
x = df_tot["year"]
y = df_tot["total"]
fit = np.polyfit(x, y, deg=1)
fit
```

```python
# Step 2: Plot the regression line on the scatter plot

df_tot.plot(kind="scatter", x="year", y="total", figsize=(10, 6), color='darkgreen')
plt.title("Total Immigration to Canada from 1980 - 2013")
plt.xlabel("Year")
plt.ylabel("Number of Immigrants")

# plot line of best fit
plt.plot(x, fit[0] * x + fit[1], color="red")
plt.annotate("y={0:.0f} x + {1:.0f}".format(fit[0], fit[1]), xy=(2000, 150000))
plt.show()

# print out the line of best fit
"No. of Immigrants = {0:.0f}*Year + {1:.0f}".format(fit[0], fit[1])
```

```python
# Bubble Plots

# compare Argentina's immigration to that of it's neighbour Brazil.
df_can_t = df_can[years].transpose()
# cast the Years (the index) to type int
df_can_t.index = map(int, df_can_t.index)
# let's label the index. This will automatically be the column name when we reset the index
df_can_t.index.name = "Year"
# reset index to bring the Year in as a column
df_can_t.reset_index(inplace=True)
df_can_t.head()
```

```python
# normalize Brazil data
norm_brazil = (df_can_t["Brazil"] - df_can_t["Brazil"].min())/(df_can_t["Brazil"].max() - df_can_t["Brazil"].min())
norm_argentina = (df_can_t["Argentina"] - df_can_t["Argentina"].min())/(df_can_t["Argentina"].max() - df_can_t["Argentina"].min())
```

```python
# Brazil
ax0 = df_can_t.plot(kind="scatter", x="Year", y="Brazil", figsize=(14,8), alpha=0.5, color="green",
                    s=norm_brazil * 2000 +10,  # pass in weights
                    xlim=(1975, 2015)
                   )
ax1 = df_can_t.plot(kind="scatter", x="Year", y="Argentina", figsize=(14,8), alpha=0.5, color="blue",
                   s=norm_argentina * 2000 +10, ax=ax0)
ax0.set_title("Immigration from Brazil and Argentina from 1980 to 2013")
ax0.set_ylabel("Number of Immigrants")
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```

```python
# Create bubble plots of immigration from China and India to visualize any differences with time from 1980 to 2013

#Step 1: Normalize the data pertaining to China and India.
norm_china = (df_can_t["China"] - df_can_t["China"].min())/(df_can_t["China"].max() - df_can_t["China"].min())
norm_india = (df_can_t["India"] - df_can_t["India"].min())/(df_can_t["India"].max() - df_can_t["India"].min())
```

```python
# Step 2: Generate the bubble plots.
ax0 = df_can_t.plot(kind="scatter", x="Year", y="China", figsize=(14,8), alpha=0.5, color="green",
                    s=norm_china * 2000 +10,  # pass in weights
                    xlim=(1975, 2015)
                   )
ax1 = df_can_t.plot(kind="scatter", x="Year", y="India", figsize=(14,8), alpha=0.5, color="blue",
                   s=norm_india * 2000 +10, ax=ax0)
ax0.set_title("Immigration from China and India from 1980 to 2013")
ax0.set_ylabel("Number of Immigrants")
ax0.legend(['China', 'India'], loc='upper left', fontsize='x-large')
```
