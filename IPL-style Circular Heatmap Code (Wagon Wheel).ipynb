```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle
import matplotlib.patches as mpatches

# Simulated IPL shot data
np.random.seed(42)
n_shots = 100

# Generate random angles and assign shot types
angles = np.random.uniform(0, 2 * np.pi, n_shots)
shot_types = np.random.choice(['Single', 'Four', 'Six'], size=n_shots, p=[0.5, 0.3, 0.2])
distance_map = {'Single': 2, 'Four': 4, 'Six': 5.5}
color_map = {'Single': 'yellow', 'Four': 'orange', 'Six': 'red'}

# Create plot
fig, ax = plt.subplots(figsize=(7, 7), subplot_kw={'projection': 'polar'})
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)
ax.set_yticklabels([])
ax.set_xticklabels([])

# Background ground
ground = Circle((0, 0), radius=7, transform=ax.transData._b, color='forestgreen', zorder=0)
ax.add_patch(ground)

# Boundary line
boundary = Circle((0, 0), radius=6.5, transform=ax.transData._b, fill=False,
                  linestyle='--', color='white', linewidth=1.5, zorder=1)
ax.add_patch(boundary)

# Plot shots
for angle, shot in zip(angles, shot_types):
    distance = distance_map[shot]
    ax.plot([angle], [distance], marker='o', markersize=10,
            color=color_map[shot], markeredgecolor='black', zorder=2)

# Draw pitch
pitch_length = 3
pitch_width = 0.4
pitch = Rectangle((-pitch_width/2, -pitch_length/2), pitch_width, pitch_length,
                  transform=ax.transData._b, facecolor='#A0522D',
                  edgecolor='peru', linewidth=2, zorder=3)
ax.add_patch(pitch)

# Creases
for y in [-pitch_length/2 + 0.2, pitch_length/2 - 0.2]:
    crease = Rectangle((-pitch_width/2, y), pitch_width, 0.06,
                       transform=ax.transData._b, color='white', zorder=4)
    ax.add_patch(crease)

# Add direction labels (approximate positions)
labels = {
    "Cover": 45, "Midwicket": 135, "Square Leg": 110,
    "Fine Leg": 160, "Third Man": 200, "Point": 315,
    "Long Off": 30, "Long On": 150
}
for label, deg in labels.items():
    angle_rad = np.deg2rad(deg)
    ax.text(angle_rad, 6.9, label, fontsize=9, ha='center', va='center', color='white')

# Add legend
legend_patches = [mpatches.Patch(color=color_map[k], label=k) for k in color_map]
ax.legend(handles=legend_patches, loc='upper center', bbox_to_anchor=(0.5, 1.08), ncol=3, fontsize=10)

# Title and player name
ax.set_title(" IPL Shot Wagon Wheel\nPlayer: Virat Kohli", fontsize=16, weight='bold', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle
import matplotlib.patches as mpatches

# Simulate 360-degree shots
np.random.seed(42)
n_directions = 72  # every 5 degrees
angles = np.linspace(0, 2 * np.pi, n_directions, endpoint=False)

# Assign shot types realistically
shot_types = np.random.choice(['Single', 'Double', 'Triple', 'Four', 'Six'], size=n_directions, 
                              p=[0.35, 0.25, 0.15, 0.15, 0.1])

# Color & distance mapping
color_map = {
    'Single': 'yellow',
    'Double': 'limegreen',
    'Triple': 'dodgerblue',
    'Four': 'orange',
    'Six': 'red'
}

# Distances (base radius is 1 unit)
distance_map = {
    'Single': 2,
    'Double': 3,
    'Triple': 4.5,
    'Four': 5.5,
    'Six': 6.5
}

# Uniform bar width and base radius
bar_width = 2 * np.pi / n_directions
radius_base = 1

# Plot setup
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': 'polar'})
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)
ax.set_yticklabels([])
ax.set_xticklabels([])

# -----------------------
# Green ground background
# -----------------------
ground = Circle((0, 0), radius=radius_base + 6.8, transform=ax.transData._b,
                color='forestgreen', zorder=0)
ax.add_patch(ground)

# Boundary ring at 6 units
boundary = Circle((0, 0), radius=radius_base + 6, transform=ax.transData._b,
                  fill=False, linestyle='--', color='white', linewidth=1.5, zorder=1)
ax.add_patch(boundary)

# -----------------------
# Shot bars (wagon arcs)
# -----------------------
for angle, shot_type in zip(angles, shot_types):
    color = color_map[shot_type]
    height = distance_map[shot_type]
    ax.bar(angle, height, width=bar_width, bottom=radius_base,
           color=color, edgecolor='black', linewidth=0.4, zorder=2)

# -----------------------
# Pitch drawing
# -----------------------
pitch_length = 3
pitch_width = 0.4
pitch = Rectangle((-pitch_width / 2, -pitch_length / 2), pitch_width, pitch_length,
                  transform=ax.transData._b, facecolor='#A0522D',
                  edgecolor='saddlebrown', linewidth=2, zorder=3)
ax.add_patch(pitch)

# Creases
for y in [-pitch_length / 2 + 0.2, pitch_length / 2 - 0.2]:
    crease = Rectangle((-pitch_width / 2, y), pitch_width, 0.06,
                       transform=ax.transData._b, color='white', zorder=4)
    ax.add_patch(crease)

# -----------------------
# Direction labels
# -----------------------
direction_labels = {
    "Cover": 45, "Midwicket": 135, "Square Leg": 110,
    "Fine Leg": 160, "Third Man": 200, "Point": 315,
    "Long Off": 30, "Long On": 150, "Straight": 0, "Behind": 180
}
for label, degree in direction_labels.items():
    angle_rad = np.deg2rad(degree)
    ax.text(angle_rad, 7, label, fontsize=9, ha='center',
            va='center', color='white', fontweight='bold', zorder=5)

# -----------------------
# Legend & Title
# -----------------------
legend_patches = [mpatches.Patch(color=color_map[k], label=k) for k in color_map]
ax.legend(handles=legend_patches, loc='upper center',
          bbox_to_anchor=(0.5, 1.1), ncol=3, fontsize=10, frameon=True)

ax.set_title(" IPL Circular Shot Heatmap\nPlayer: Virat Kohli vs MI", fontsize=16, weight='bold', va='bottom')

plt.tight_layout()
plt.show()
```
