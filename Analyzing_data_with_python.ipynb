```python
import pandas as pd
```

```python
import sqlite3
```

```python
data = pd.read_csv(r"F:\Data_Science_Jupyter_Files\menu.csv")
```

```python
connection = sqlite3.connect("McDonalds.db")
```

```python
data.to_sql("MCDONALDS_NUTRITION", connection)
```

```python
df=pd.read_sql("SELECT * FROM MCDONALDS_NUTRITION", connection)
```

Which food item has maximum sodium content?

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import seaborn as sns
```

```python
#categorical scatter plot
plot = sns.swarmplot(x="Category", y="Sodium", data=df)
plt.setp(plot.get_xticklabels(), rotation = 70)
plt.title("Sodium Content")
plt.show()
```
